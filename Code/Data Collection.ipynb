```python
import os.path
from Katna.image import Image
import os
import ntpath


def main():

    img_module = Image()

    # folder to save resized images
    output_folder_resized_image = '/content/drive/MyDrive/Embedded AI Project /Test/files'#"resizedimages"

    if not os.path.isdir(os.path.join(".", output_folder_resized_image)):
        os.mkdir(os.path.join(".", output_folder_resized_image))

    # resized image dimensions
    resize_width = 96
    resize_height = 96

    # Input folder file path
    # input_folder_path = os.path.join(".", "tests", "data")
    # print(f"input_folder_path = {input_folder_path}")

    resized_images = img_module.resize_image_from_dir(
        dir_path='/content/drive/MyDrive/Embedded AI Project /Test/files',
        target_width=resize_width,
        target_height=resize_height,
        down_sample_factor=8,
    )

    for filepath, resized_image in resized_images.items():
        # name of the image file
        filename = ntpath.basename(filepath)
        name = filename.split(".")[0]
        # folder path where the images will be stored
        img_module.save_image_to_disk(
            resized_image, output_folder_resized_image, name + "_resized" + "_", ".jpeg"
        )


main()
```

```python
import os
import cv2
import numpy as np

dataset_dir = "/content/drive/MyDrive/Embedded AI Project /Images_of_Waste/YOLO_imgs"

data = []
labels = []

for filename in os.listdir(dataset_dir):
    if filename.endswith(".jpg"):
        # read the image and resize it to a fixed size
        image = cv2.imread(os.path.join(dataset_dir, filename))
        image = cv2.resize(image, (224, 224))

        # read the label from the corresponding .txt file
        label_filename = os.path.splitext(filename)[0] + ".txt"
        with open(os.path.join(dataset_dir, label_filename)) as f:
            label = int(f.readline().strip()[0])
        data.append(image)
        labels.append(label)

data = np.array(data, dtype="float32")
labels = np.array(labels)

data /= 255.0
```
